# 1) Faça um crawler que navegue pelas páginas de países e baixe os htmls.

```python
from bs4 import BeautifulSoup
import requests

# Faz a solicitação do site
url = 'http://127.0.0.1:8000/places/default/index'
response = requests.get(url)

# Extrai o html
soup = BeautifulSoup(response.content)

print(soup.prettify())
```

```python
# Cria uma lista com os nomes dos paises
country_names = [name.text.strip() for name in soup.body.table.find_all('a')]

# Cria uma lista com os links para a página dos países
links_list = list(map(lambda x: x.get('href'), soup.body.section.find_all('a')))[0:-1]

# Cria um dicionário com o nome dos países, um número de referência e seus links
pages = {}
for idx, name in enumerate(country_names):
    pages.update({name: [idx, links_list[idx]]})


display(pages)
```

```python
# Cria uma lista de objetos BeautifulSoup com o html de cada país
html_countries = []
for name, link in pages.items():
    html_countries.append(BeautifulSoup(requests.get('http://127.0.0.1:8000' + link[1]).content))


# EX: Conteúdo html do país Afghanistan
html_countries[0]
```
