```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import random
import math

X, _ = make_blobs(n_samples=1000, centers=2, cluster_std=1.5)

x1 = []
x2 = []

for x in X:
    x1.append(tuple(x)[0])
    x2.append(tuple(x)[1])

plt.scatter(x=x1[:500], y=x2[:500], s=2)
plt.savefig('1.png')
plt.show()


# inicializao dos centrides
c1 = (random.randint(-15, 15), random.randint(-15, 15))
c2 = (random.randint(-15, 15), random.randint(-15, 15))

# criao dos clusters vazios (sem pontos)
cluster_c1 = []
cluster_c2 = []


for i in range(0, 5):
    P = (x1[i], x2[i])

    # calcula a distancia do ponto em relao ao centroide c1
    c1_distance = math.sqrt(math.pow(P[0] - c1[0], 2) + math.pow(P[1] - c1[1], 2))
    c2_distance = math.sqrt(math.pow(P[0] - c2[0], 2) + math.pow(P[1] - c2[1], 2))
    
    if c1_distance > c2_distance:
        cluster_c2.append(P)
    else:
        cluster_c1.append(P)


print('cluster 1', cluster_c1, '\n')

print('cluster 2', cluster_c2, '\n')

# P1
plt.scatter(x= x1[0], y= x2[0], c='blue')
plt.text(x= x1[0] + 0.2, y= x2[0] + 0.2, c='blue', s='P1')

# P2
plt.scatter(x= x1[1], y= x2[1], c='blue')
plt.text(x= x1[1] + 0.2, y= x2[1] + 0.2, c='blue', s='P2')

# P3
plt.scatter(x= x1[2], y= x2[2], c='blue')
plt.text(x= x1[2] + 0.2, y= x2[2] + 0.2, c='blue', s='P3')

# P4
plt.scatter(x= x1[3], y= x2[3], c='blue')
plt.text(x= x1[3] + 0.2, y= x2[3] + 0.2, c='blue', s='P4')

# P5
plt.scatter(x= x1[4], y= x2[4], c='blue')
plt.text(x= x1[4] + 0.2, y= x2[4] + 0.2, c='blue', s='P5')

# centroids
plt.scatter(x=c1[0], y=c1[1], c='red')
plt.text(x=c1[0] + 0.2, y=c1[1] + 0.2, c='red', s='C1')
plt.scatter(x=c2[0], y=c2[1], c='red')
plt.text(x=c2[0] + 0.2, y=c2[1] + 0.2, c='red', s='C2')

plt.plot()
plt.show()



for i in range(0, 5):
    P = (x1[i], x2[i])

    # calcula a distancia do ponto em relao ao centroide c1
    c1_distance = math.sqrt(math.pow(P[0] - c1[0], 2) + math.pow(P[1] - c1[1], 2))
    c2_distance = math.sqrt(math.pow(P[0] - c2[0], 2) + math.pow(P[1] - c2[1], 2))
    
    if c1_distance > c2_distance:
        cluster_c2.append(P)
    else:
        cluster_c1.append(P)


# atualiza a posio de c1
sum_x1 = 0
sum_x2 = 0
for i in range(len(cluster_c1)):
    sum_x1 = sum_x1 + cluster_c1[i][0]
    sum_x2 = sum_x2 + cluster_c1[i][1]
print(c1, (sum_x1, sum_x2))

# atualiza a posio de c2
sum_x1 = 0
sum_x2 = 0
for i in range(len(cluster_c2)):
    sum_x1 = sum_x1 + cluster_c2[i][0]
    sum_x2 = sum_x2 + cluster_c2[i][1]

c1 = (sum_x1 / len(cluster_c1), sum_x2 / len(cluster_c1))
c2 = (sum_x1 / len(cluster_c2), sum_x2 / len(cluster_c2))

# P1
plt.scatter(x= x1[0], y= x2[0], c='blue')
plt.text(x= x1[0] + 0.2, y= x2[0] + 0.2, c='blue', s='P1')

# P2
plt.scatter(x= x1[1], y= x2[1], c='blue')
plt.text(x= x1[1] + 0.2, y= x2[1] + 0.2, c='blue', s='P2')

# P3
plt.scatter(x= x1[2], y= x2[2], c='blue')
plt.text(x= x1[2] + 0.2, y= x2[2] + 0.2, c='blue', s='P3')

# P4
plt.scatter(x= x1[3], y= x2[3], c='blue')
plt.text(x= x1[3] + 0.2, y= x2[3] + 0.2, c='blue', s='P4')

# P5
plt.scatter(x= x1[4], y= x2[4], c='blue')
plt.text(x= x1[4] + 0.2, y= x2[4] + 0.2, c='blue', s='P5')

# centroids
plt.scatter(x=c1[0], y=c1[1], c='red')
plt.text(x=c1[0] + 0.2, y=c1[1] + 0.2, c='red', s='C1')
plt.scatter(x=c2[0], y=c2[1], c='red')
plt.text(x=c2[0] + 0.2, y=c2[1] + 0.2, c='red', s='C2')

plt.plot()
plt.show()
```
