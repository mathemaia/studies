```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def comparar_fotos(imagem1, imagem2):
    # Carregar as imagens
    img1 = cv2.imread(imagem1, cv2.IMREAD_GRAYSCALE)
    img2 = cv2.imread(imagem2, cv2.IMREAD_GRAYSCALE)

    # Redimensionar as imagens para o mesmo tamanho, se necessário
    if img1.shape != img2.shape:
        img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    # Inicializar o detector SIFT
    sift = cv2.SIFT_create()

    # Detectar e descrever características nas imagens
    keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
    keypoints2, descriptors2 = sift.detectAndCompute(img2, None)

    # Inicializar o algoritmo de correspondência de pontos (Brute-Force)
    matcher = cv2.BFMatcher()

    # Encontrar correspondências entre os descritores das imagens
    matches = matcher.match(descriptors1, descriptors2)

    # Ordenar as correspondências pela distância
    matches = sorted(matches, key=lambda x: x.distance)

    # Calcular a similaridade entre as imagens (quantidade de correspondências)
    similaridade = len(matches)

    return similaridade

# Exemplo de uso
imagem1 = 'data/nolabel.png'
imagem2 = 'data/wrong.jpg'

coeficiente = comparar_fotos(imagem1, imagem2)
print(f"corr:       {coeficiente}")


img1 = cv2.imread(imagem1, cv2.IMREAD_COLOR)
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1_rgb)
plt.show()

img2 = cv2.imread(imagem2, cv2.IMREAD_COLOR)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2_rgb)
plt.show()
```
