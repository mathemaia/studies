```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os
from sklearn.cluster import KMeans
```

```python
image = imread('R.jpeg')
X = image.reshape(-1, 3)
kmeans = KMeans(n_clusters=8).fit(X)

segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)

segmented_img = segmented_img / 255
plt.imshow(segmented_img)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits

mnist = load_digits()
X = mnist.data
y = mnist.target

kmeans = KMeans(n_clusters=10)  # Definir o nmero de clusters desejado
kmeans.fit(X)
labels = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definir o nmero de clusters desejado
n_clusters = 10

# Criar uma matriz de subplots para exibir as imagens dos clusters
fig, axs = plt.subplots(n_clusters, 10, figsize=(10, n_clusters))

# Iterar sobre os clusters
for cluster_label in range(n_clusters):
    # Obter os ndices das amostras pertencentes ao cluster atual
    cluster_indices = np.where(labels == cluster_label)[0]
    
    # Iterar sobre as primeiras 10 amostras do cluster atual
    for i, sample_index in enumerate(cluster_indices[:10]):
        # Obter a imagem correspondente ao ndice da amostra
        image = X[sample_index].reshape(8, 8)
        
        # Plotar a imagem no subplot correspondente
        axs[cluster_label, i].imshow(image, cmap='gray')
        axs[cluster_label, i].axis('off')

# Ajustar o espaamento entre os subplots
plt.tight_layout()

# Exibir o plot
plt.show()
```
